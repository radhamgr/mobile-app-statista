# Removing Duplicate Entries

```python
for app in android:
    name = app[0]
    if name == 'Instagram':
        print(app)
```

```python
duplicate_app = []
unique_app = []
for app in android:
    name = app[0]
    if name in unique_app:
        duplicate_app.append(name) 
    else:
        unique_app.append(name)
        
print('Number of duplicate apps:', len(duplicate_app))
print(len(unique_app))
print('\n')
print('Examples of duplicate apps:', duplicate_app[:15])
```

```python
reviews_max = {}
for app in android:
    name = app[0]
    n_reviews = float(app[3])
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
        
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
        
    
```

```python
print('Expected length:', len(android) - 1181)
print('Actual length:', len(reviews_max))
```

```python
android_clean = []
already_clean = []

for app in android:
    name = app[0]
    n_reviews = float(app[3])
    
    if (reviews_max[name] == n_reviews) and (name not in already_clean)  :
        android_clean.append(app)
        already_clean.append(name)
```

# Most Common Apps by Genre

```python
def freq_table(dataset,index):
    table = {}
    total = 0
    for row in dataset:
        total += 1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1 
            
    table_percentages ={}
    for key in table:
        percentage = (table[key]/total) * 100
        
        table_percentages[key] = percentage 
    return table_percentages
```

```python
freq_table(android_final,1)
```

```python
def display_table(dataset,index):
    table_display = []
    table = freq_table(dataset,index)
    for keys in table:
        key_val = (table[keys],keys)
        table_display.append(key_val)
        
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
    
        
    
```

```python
display_table(android_final,1)
```

```python
display_table(ios_final, -5)
```

# Most Popular Apps by Genre on the App Store

```python
genres_ios = freq_table(ios_final, -5)

for genre in genres_ios:
    total = 0
    len_genre = 0
    for app in ios_final:
        genre_app = app[-5]
        if genre_app == genre:            
            n_ratings = float(app[5])
            total += n_ratings
            len_genre += 1
    avg_n_ratings = total / len_genre
    print(genre, ':', avg_n_ratings)
```
